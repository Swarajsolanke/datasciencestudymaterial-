```python
# first task is to divide the dataset into independent and depend set
X=df.iloc[:,:-1] # independent feature 
y=df.iloc[:,-1]# depend feature 
```

```python
print(X)
```

```python
print(y) # target feature 
```

```python
from sklearn.model_selection import train_test_split
```

```python
#here i split the data into test and train
X_train, X_test, y_train, y_test = train_test_split(
     X, y, test_size=0.30, random_state=36)
```

```python
X_train.shape,X_test.shape 
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler.fit(X_train) # here i calculated the mean and standard deviation 
```

```python
X_train=scaler.fit_transform(X_train) #here i calculated the (z-score =x-u/sigma)
```

```python
X_train
```

```python
X_test=scaler.fit_transform(X_test) # here only calculating the z-score 
```

```python
from sklearn.linear_model import LinearRegression #here i imported the linear regression model
```

```python
regression=LinearRegression()
```

```python
regression.fit(X_train,y_train) # here i passed the x_train and y_train data 
```

```python
regression.coef_ # here i find the coefficient .
```

```python
regression.intercept_ #when your all inputs are zero then this is all you will get
```

```python
## prediction
y_pred=regression.predict(X_test)
```

```python
# here we find the means sqaured and mean absolute error 
from sklearn.metrics import mean_squared_error,mean_absolute_error
```

```python
import numpy as np
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
## Accuracy r2 and adjusted r square
from sklearn.metrics import r2_score
```

```python
score=r2_score(y_test,y_pred)
```

```python
score 
```

```python
#display adjusted R-squared
1 - (1-score)*(len(y)-1)/(len(y)-X.shape[1]-1)
```

```python
from sklearn.linear_model import Ridge
ridge=Ridge(alpha=20.0)
ridge.fit(X_train,y_train)
```

```python
y_pred=ridge.predict(X_test)
```

```python
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import Lasso
```

```python
lasso=Lasso(alpha=20.0)
lasso.fit(X_train,y_train)
```

```python
y_pred=lasso.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```

```python
from sklearn.linear_model import ElasticNet
```

```python
y_pred=elasticnet.predict(X_test)
mse=mean_squared_error(y_test,y_pred)
print(mse)
mae=mean_absolute_error(y_test,y_pred)
print(mae)
print(np.sqrt(mse))
```
