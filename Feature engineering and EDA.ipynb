# onehotencoder

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
ohe.fit_transform(df[['Airline']]).toarray() # we have to fit the data and then transform it
```

```python
## Replacing target guided ordinal encoding
def replace_airline_with_mean(df): # this fuction convert the categorical variable into mean value base on the output feature 
    mean_prices = df.groupby('Airline')['Price'].mean().sort_values()
    df['Airline'] = df['Airline'].apply(lambda x: mean_prices[x])
    return df

df = replace_airline_with_mean(df)
df.head()
```
